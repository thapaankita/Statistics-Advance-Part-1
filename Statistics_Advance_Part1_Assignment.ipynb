#Basic Question

## Q1. What is a random variable in probability theory?

### Answer:

A **random variable** is a function or rule that assigns a numerical value to each possible outcome of a random experiment. It provides a way to describe uncertain events in a mathematical form, making it easier to analyze and compute probabilities, expectations, and other statistical measures.

In probability theory, a random variable acts as a bridge between real-world random processes (such as tossing a coin or measuring rainfall) and mathematical models. Instead of dealing with uncertain events directly, we work with the numerical values assigned by the random variable.

## Q2. What are the types of random variables?

### Answer:

In probability theory, random variables are classified into two main types based on the nature of the values they can take:

### **1. Discrete Random Variable:**

A discrete random variable can take on a countable number of distinct values. These values are often integers, and there are gaps between the possible values.

### **2. Continuous Random Variable:**

A continuous random variable can take on an infinite number of possible values within a given range. The values are uncountable and can include decimals or fractions.

## Q3. What is the difference between discrete and continuous distributions?

### Answer:

The key difference between **discrete** and **continuous** distributions lies in the type of values that the random variable can take and how those values are represented through probability.

### **1. Discrete Distribution:**

A discrete distribution describes the probability of outcomes for a **discrete random variable**, which takes on a finite or countably infinite set of distinct values.

**Characteristics:**
- The possible outcomes are countable.
- Each individual outcome has a specific probability.
- The sum of all probabilities is exactly 1.

**Examples:**
- Binomial Distribution (e.g., number of successes in a series of trials).
- Poisson Distribution (e.g., number of emails received per hour).
- Probability of rolling each number on a six-sided die.

### **2. Continuous Distribution:**

A continuous distribution describes the probability for a **continuous random variable**, which can take any value within a certain range or interval.

**Characteristics:**
- The possible outcomes are uncountable (infinite within a range).
- Probabilities are represented using a probability density function (PDF).
- The probability of the variable taking any exact value is technically zero; instead, we calculate the probability over an interval.
- The total area under the PDF curve is 1.

**Examples:**
- Normal Distribution (e.g., heights of people).
- Uniform Distribution (e.g., random number between 0 and 1).
- Exponential Distribution (e.g., time between events).

## Q4. What are Probability Distribution Functions (PDF)?

### Answer:

A **Probability Distribution Function (PDF)** describes how the probabilities are distributed over the possible values of a **continuous random variable**.

In other words, for continuous random variables, the PDF is a mathematical function that defines the likelihood of the variable falling within a specific range of values.

### **Key Points about PDFs:**

- The PDF is denoted as `f(x)` or `p(x)`.
- The area under the curve of the PDF between two values gives the probability that the random variable falls within that range.
- The probability that the random variable takes an exact value is technically zero.
- The total area under the entire PDF curve is always equal to **1**, representing the certainty that some outcome will occur.

## Q5. How do Cumulative Distribution Functions (CDF) differ from Probability Distribution Functions (PDF)?

### Answer:

The **Probability Distribution Function (PDF)** and the **Cumulative Distribution Function (CDF)** are both used to describe the behavior of a random variable, but they serve different purposes.

### **Probability Distribution Function (PDF):**

The PDF describes the **likelihood** of a continuous random variable falling within a specific range of values. It provides a curve that shows how likely different outcomes are, but the probability of any exact value is technically zero.

For continuous random variables:

\[
P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
\]

Where:
- `f(x)` is the PDF.
- The area under the curve between `a` and `b` gives the probability that `X` falls within that range.

### **Cumulative Distribution Function (CDF):**

The CDF gives the **probability** that a random variable takes a value **less than or equal to** a specific number. It accumulates the probabilities from the lower bound up to that number.

The CDF is defined as:

\[
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
\]

Where:
- `F(x)` is the CDF.
- `f(t)` is the corresponding PDF.
- The CDF always increases as `x` increases and ranges from 0 to 1.

## Q6. What is a Discrete Uniform Distribution?

### Answer:

A **Discrete Uniform Distribution** is a type of probability distribution in which a finite number of distinct outcomes are all equally likely to occur. This means that every possible value of the random variable has the same constant probability assigned to it.

In other words, if a random experiment results in any one of `n` possible distinct outcomes, and there is no preference or bias towards any particular outcome, then the distribution of those outcomes is considered discrete and uniform.

Mathematically, if a random variable `X` takes on `n` possible distinct values `{x₁, x₂, x₃, ..., xₙ}`, the probability of each outcome is given by:

\[
P(X = x_i) = \frac{1}{n} \quad \text{for} \quad i = 1, 2, 3, ..., n
\]

Since all outcomes are equally likely, the sum of the probabilities for all possible outcomes will always be 1.

The discrete uniform distribution is one of the simplest probability models and is often used to represent situations where all outcomes are equally likely, such as rolling a fair die, drawing a card from a well-shuffled deck, or randomly selecting a number from a specific set of integers.

## Q7. What are the key properties of a Bernoulli Distribution?

### Answer:

The **Bernoulli Distribution** is one of the simplest and most fundamental probability distributions in statistics. It describes a random experiment that has exactly two possible outcomes: **success** (usually denoted by `1`) and **failure** (usually denoted by `0`). The Bernoulli distribution is widely used to model binary events where only two outcomes are possible.

The distribution is completely characterized by a single parameter `p`, where:
- `p` is the probability of success (`X = 1`).
- `(1 - p)` is the probability of failure (`X = 0`).

The key properties of a Bernoulli distribution are as follows:

1. **Two Possible Outcomes:**
   - The random variable `X` can only take the values `0` (failure) or `1` (success).

2. **Probability Mass Function (PMF):**
   The probability for each possible outcome is given by:
   
   \[
   P(X = x) = p^x (1 - p)^{1 - x} \quad \text{for} \quad x = 0 \ \text{or} \ 1
   \]
   
   Where:
   - `p` is the probability of success (`0 ≤ p ≤ 1`).
   - `(1 - p)` is the probability of failure.

3. **Mean (Expected Value):**
   
   \[
   E(X) = p
   \]
   
   The expected value of a Bernoulli random variable is simply the probability of success.

4. **Variance:**
   
   \[
   Var(X) = p(1 - p)
   \]
   
   The variance depends on both the probability of success and failure.

5. **Applications:**
   - The Bernoulli distribution is used to model simple yes/no or true/false experiments, such as:
     - Tossing a coin (Heads = success, Tails = failure).
     - Passing or failing a test.
     - Defective or non-defective product in quality control.

The Bernoulli distribution serves as the building block for more complex distributions, such as the Binomial distribution, which models the number of successes in multiple independent Bernoulli trials.

## Q8. What is the Binomial Distribution, and how is it used in Probability?

### Answer:

The **Binomial Distribution** is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a binary (yes/no) experiment, where each trial has only two possible outcomes: **success** or **failure**.

In a binomial experiment:
- Each trial is independent, meaning the outcome of one trial does not affect the others.
- The probability of success, denoted by `p`, remains constant for each trial.
- The probability of failure is `(1 - p)`.

The binomial distribution is widely used to model situations where we are interested in counting the number of times a particular outcome (success) occurs over a fixed number of repeated trials.

### **Usage in Probability:**

The binomial distribution is used to model and solve problems involving:
- The number of heads obtained in a fixed number of coin tosses.
- The number of defective items found in a sample of products.
- The number of students passing an exam if each has the same probability of passing.
- Any real-world situation involving a fixed number of independent trials with two possible outcomes.

By using the binomial distribution, we can calculate the likelihood of obtaining a certain number of successes and make informed decisions based on probabilistic reasoning.

## Q9. What is the Poisson Distribution and where is it applied?

### Answer:

The **Poisson Distribution** is a discrete probability distribution that describes the probability of a certain number of events happening within a fixed period of time, distance, area, or volume, provided that the events occur independently and at a constant average rate.

It is particularly useful for modeling situations where events happen randomly, unpredictably, and infrequently, but with a known average frequency. The Poisson distribution does not concern itself with the exact timing of the events, but rather with the total number of events occurring within the given interval.

### **Applications of Poisson Distribution:**

The Poisson distribution is widely used to model real-world scenarios where:
- Events happen independently of each other.
- The average rate at which events occur is constant over time or space.
- The probability of multiple events happening at the exact same moment is very small.

**Common examples of where the Poisson distribution is applied include:**
- The number of customer arrivals at a store or service center within an hour.
- The number of emails or phone calls received in a specific time frame.
- The number of accidents happening at a traffic intersection in a day.
- The number of printing errors on a page of a book.
- The number of defects found in a fixed length of manufacturing material.

## Q10. What is a Continuous Uniform Distribution?

### Answer:

A **Continuous Uniform Distribution** is a type of probability distribution in which all outcomes within a certain continuous range are equally likely to occur. In other words, for a continuous random variable, every value within the specified interval has the same constant probability density, and any value outside that interval has zero probability.

This distribution is often referred to as a **rectangular distribution** because its probability density function forms a rectangle when graphed.

## Q11. What are the Characteristics of a Normal Distribution?

### Answer:

The **Normal Distribution**, also known as the **Gaussian Distribution**, is one of the most important and widely used probability distributions in statistics and probability theory. It describes a continuous, symmetrical, bell-shaped curve where most of the data points cluster around the mean, and the probability of extreme values decreases as you move further from the mean.

Many real-world phenomena, such as heights, weights, test scores, and measurement errors, tend to follow a normal distribution under appropriate conditions.

### **Key Characteristics of a Normal Distribution:**

1. **Symmetry:**
   - The normal distribution is perfectly symmetric around its mean.
   - The left and right halves of the curve are mirror images of each other.

2. **Bell-Shaped Curve:**
   - The curve has a single peak at the mean.
   - It is highest at the center and tapers off smoothly towards both ends.

3. **Mean, Median, and Mode are Equal:**
   - In a normal distribution, the mean, median, and mode all have the same value, located at the center of the distribution.

4. **Defined by Two Parameters:**
   - The distribution is completely described by:
     - The **mean (μ)**, which determines the center of the curve.
     - The **standard deviation (σ)**, which controls the spread or width of the curve.

5. **68-95-99.7 Rule (Empirical Rule):**
   - About **68%** of the data falls within **1 standard deviation** from the mean.
   - About **95%** falls within **2 standard deviations**.
   - About **99.7%** falls within **3 standard deviations**.

6. **Asymptotic Nature:**
   - The curve approaches, but never actually touches, the horizontal axis as it extends infinitely in both directions.

7. **Unimodal:**
   - The curve has only one peak, which occurs at the mean.

8. **Area Under the Curve:**
   - The total area under the curve is exactly 1, representing the total probability.

## Q12. What is the Standard Normal Distribution, and why is it important?

### Answer:

The **Standard Normal Distribution** is a special case of the normal distribution where the mean is exactly **0** and the standard deviation is exactly **1**. Like all normal distributions, it follows a symmetrical, bell-shaped curve, but its specific parameters (mean = 0 and standard deviation = 1) make it a standardized version used widely in statistics.

The standard normal distribution is used to simplify calculations and allow for easy comparison across different datasets that may originally have different means and standard deviations.

### **Importance of the Standard Normal Distribution:**

1. **Basis for Z-Scores:**
   - The standard normal distribution forms the foundation for calculating **z-scores**, which measure how many standard deviations a particular data point is away from the mean.
   - Z-scores help standardize different datasets and allow for direct comparison, regardless of the original scale.

2. **Simplifies Probability Calculations:**
   - By converting values to z-scores, we can use standard normal tables (or software) to easily find probabilities and percentiles without complex calculations for each individual dataset.

3. **Widely Used in Hypothesis Testing:**
   - Many statistical tests, such as the z-test, rely on the standard normal distribution to determine critical values and p-values for decision-making.

4. **Foundation for Statistical Inference:**
   - The standard normal distribution is crucial for estimating confidence intervals, conducting hypothesis tests, and making probabilistic statements about populations.

5. **Universally Applicable Through Standardization:**
   - Any normal distribution, regardless of its mean or standard deviation, can be converted to the standard normal distribution through a simple transformation process called **standardization**, making the distribution universally applicable.

6. **Visual Simplicity:**
   - The standard normal curve provides a clear and consistent reference for understanding probabilities, proportions, and statistical significance.

## Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

### Answer:

The **Central Limit Theorem (CLT)** is one of the most important and powerful concepts in statistics. It states that, regardless of the original distribution of a population, the distribution of the sample means will tend to approximate a **normal distribution** as the sample size becomes large, provided the samples are independent and identically distributed.

In simpler terms, even if the population data is not normally distributed, the distribution of the averages (means) of large enough samples drawn from that population will follow a normal distribution.

### Importance of the Central Limit Theorem:

- It allows statisticians to apply **normal distribution-based techniques** even when the population is not normally distributed.
- It provides the theoretical foundation for many statistical methods such as confidence intervals, hypothesis testing, and control charts.
- It simplifies complex problems by enabling the use of z-scores and probability calculations based on the normal distribution.
- It plays a crucial role in fields like quality control, economics, medical research, and social sciences where sampling is commonly used.

## Q14. How does the Central Limit Theorem relate to the normal distribution?

### Answer:

The **Central Limit Theorem (CLT)** is directly related to the normal distribution because it explains how and why the normal distribution frequently appears in statistical analysis, even when the original population distribution is not normal.

According to the CLT, when independent random samples of a sufficiently large size are drawn from any population, regardless of the population's original distribution, the distribution of the sample means will approximate a **normal distribution**.

This relationship holds true under the following conditions:
- The samples are drawn randomly and independently.
- The sample size is sufficiently large (commonly, 30 or more observations are considered adequate).
- The population has a finite mean and variance.

## Q15. What is the application of Z statistics in hypothesis testing?

### Answer:

**Z statistics**, also known as the **Z-test**, is a statistical technique used in hypothesis testing to determine whether there is a significant difference between sample statistics and population parameters, or between two sample statistics, under the assumption that the population follows a normal distribution or that the sample size is large enough for the Central Limit Theorem to apply.

The Z-test uses a standardized score called the **Z-score**, which measures how many standard deviations a data point or sample statistic is away from the population mean. It helps in determining the probability of observing a result as extreme as the one obtained, assuming the null hypothesis is true.

### **Application of Z Statistics in Hypothesis Testing:**

1. **Testing Population Means:**
   - Z-tests are used to compare a sample mean to a known population mean when the population standard deviation is known.
   - This helps determine whether the observed sample provides enough evidence to reject the null hypothesis about the population mean.

2. **Testing Proportions:**
   - Z-tests are applied to test hypotheses about population proportions, especially in large-sample scenarios such as surveys or experiments with binary outcomes (e.g., success/failure).

3. **Comparison of Two Means or Proportions:**
   - Z-tests can be used to compare the means or proportions of two independent samples to assess whether the difference between them is statistically significant.

4. **Determining Statistical Significance:**
   - By calculating the Z-score and comparing it to critical values from the standard normal distribution, researchers can assess whether to reject or fail to reject the null hypothesis.
   - The Z-test provides the p-value, which indicates the probability of observing a result at least as extreme as the one obtained, assuming the null hypothesis is true.

5. **Widely Used in Large Sample Situations:**
   - Z-tests are particularly useful when dealing with large sample sizes, where the sampling distribution of the sample mean or proportion can be assumed to be approximately normal, even if the original population is not normally distributed.

6. **Foundation for Confidence Intervals:**
   - Z statistics are also used to construct confidence intervals for population parameters such as the mean or proportion, helping estimate the range of plausible values for those parameters.

## Q16. How do you calculate a Z-score, and what does it represent?

### Answer:

A **Z-score**, also known as a **standard score**, is a statistical measure that describes a data point's position relative to the mean of a dataset, expressed in terms of standard deviations. It indicates how many standard deviations a particular value is above or below the mean of the distribution.

The Z-score is a crucial concept in statistics because it allows for the standardization of different datasets, making them comparable, and helps in understanding the relative standing of a particular observation within a dataset.

### **How to Calculate a Z-score:**

The Z-score for a data point is calculated using the following formula:

**Z = (X - μ) / σ**

Where:
- **Z** = Z-score
- **X** = Value of the data point
- **μ** = Mean of the population or sample
- **σ** = Standard deviation of the population or sample

## Q17. What are point estimates and interval estimates in statistics?

### Answer:

In statistics, **point estimates** and **interval estimates** are methods used to estimate unknown population parameters based on sample data.

### **Point Estimates:**

A **point estimate** provides a single, specific value as the best guess for an unknown population parameter. It is calculated directly from the sample.

**Examples:**
- Sample mean estimates population mean.
- Sample proportion estimates population proportion.

Point estimates are simple but do not show how much uncertainty is involved in the estimate.

### **Interval Estimates:**

An **interval estimate** gives a range of values within which the true population parameter is likely to fall, along with a confidence level.

**Example:**
- A 95% confidence interval for the population mean might be (45, 55), meaning we are 95% confident that the true mean lies within this range.

Interval estimates provide a clearer picture of uncertainty and reliability in the estimation process.

## Q18. What is the significance of confidence intervals in statistical analysis?

### Answer:

**Confidence intervals** are important tools in statistical analysis that provide a range of values within which the true population parameter is likely to fall, based on sample data, along with an associated level of confidence (commonly 90%, 95%, or 99%).

### **Significance of Confidence Intervals:**

1. **Measure of Uncertainty:**
   - Confidence intervals account for the variability in sample data and provide a realistic range for the unknown parameter.

2. **Better Than Point Estimates Alone:**
   - Unlike point estimates, confidence intervals offer additional information about the possible error or uncertainty in the estimate.

3. **Basis for Decision Making:**
   - They help researchers and analysts make more informed decisions by showing the degree of precision in their estimates.

4. **Interpretation of Statistical Results:**
   - Confidence intervals indicate how reliable the results are and how much the estimate might vary if the study were repeated.

5. **Application Across Fields:**
   - Used in research, surveys, quality control, and experiments to express the reliability of estimated values like means or proportions.

## Q19. What is the relationship between a Z-score and a confidence interval?

### Answer:

The **Z-score** and **confidence interval** are closely related concepts in statistics, both used to estimate and interpret population parameters based on sample data.

### **Relationship:**

1. **Z-score Defines the Confidence Level:**
   - The Z-score determines how many standard deviations away from the mean are required to capture a certain percentage of the data in a normal distribution.
   - For example, a Z-score of **1.96** corresponds to a **95% confidence level**, meaning 95% of the data falls within 1.96 standard deviations from the mean.

2. **Construction of Confidence Intervals:**
   - Confidence intervals are calculated using the Z-score that matches the desired confidence level.
   - The general formula for a confidence interval is:

     **Confidence Interval = Point Estimate ± (Z-score × Standard Error)**

3. **Direct Link Through the Normal Distribution:**
   - The Z-score is based on the standard normal distribution, which forms the basis for determining how wide a confidence interval should be for a given confidence level.

## Q21. What are the assumptions for applying the Central Limit Theorem?

### Answer:

The **Central Limit Theorem (CLT)** describes how the sampling distribution of the sample mean approaches a normal distribution as the sample size increases. To apply the CLT properly, certain assumptions must be satisfied:

### **Assumptions for CLT:**

1. **Random Sampling:**
   - The samples should be drawn randomly from the population.

2. **Independence:**
   - The sampled observations must be independent of each other.
   - In practical terms, this means one observation does not influence another.

3. **Identically Distributed:**
   - All samples should come from the same population with the same distribution.

4. **Sample Size is Large Enough:**
   - A sufficiently large sample size is required for the CLT to apply.
   - Typically, a sample size of **30 or more** is considered adequate.
   - For highly skewed or non-normal populations, an even larger sample may be necessary.

5. **Finite Variance and Mean:**
   - The population from which the samples are drawn must have a finite mean and variance.

## Q22. What is the concept of expected value in a probability distribution?

### Answer:

The **expected value** in a probability distribution represents the long-run average or mean value of a random variable over many repetitions of an experiment. It provides a measure of the center or average outcome that one would expect if the random process were repeated a large number of times.

### **Explanation:**

- The expected value is a weighted average of all possible values that a random variable can take, where the weights are the probabilities of each outcome.
- It gives a theoretical measure of the "average" result, even if that exact value may not occur in every trial.

## Q2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the parameters for the discrete uniform distribution
low = 1    # Lower bound (inclusive)
high = 7   # Upper bound (exclusive) - simulating a fair die roll (1 to 6)

# Generate random samples
samples = randint.rvs(low, high, size=1000)

# Define possible outcomes
x = np.arange(low, high)

# Calculate PMF
pmf = [randint.pmf(k, low, high) for k in x]

# Plot the PMF
plt.stem(x, pmf, basefmt=" ")
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

## Q3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
from scipy.stats import bernoulli

# Define a function to calculate PDF of Bernoulli distribution
def bernoulli_pdf(p, x):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    p : float : Probability of success (0 <= p <= 1)
    x : int : Outcome (0 or 1)

    Returns:
    float : Probability of observing outcome x
    """
    return bernoulli.pmf(x, p)

# Example usage:
p = 0.6  # Probability of success
for x in [0, 1]:
    print(f"P(X = {x}) =", bernoulli_pdf(p, x))
```

## Q5. Create a Poisson distribution and visualize it using Python.

## Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1    # Lower bound (inclusive)
high = 7   # Upper bound (exclusive) - simulating a fair die (1 to 6)

# Define possible outcomes
x = np.arange(low, high)

# Calculate CDF for each outcome
cdf = [randint.cdf(k, low, high) for k in x]

# Plot the CDF
plt.step(x, cdf, where='post', color='purple', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

## Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Function to calculate Z-scores and plot them
def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores for the given dataset and plots them.

    Parameters:
    data : array-like : The dataset
    """
    z_scores = zscore(data)

    # Plot the Z-scores
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()

    return z_scores

# Example dataset
data = np.random.normal(50, 10, 100)

# Call the function
z_scores = calculate_and_plot_zscores(data)
```

## Q10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 30       # Size of each sample
num_samples = 1000     # Number of samples

# Generate a non-normal population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=10000)

# List to store sample means
sample_means = []

# Draw multiple samples and calculate their means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True, alpha=0.7)
plt.title('Demonstration of Central Limit Theorem (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

## Q12.  Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate and plot the standard normal distribution
def plot_standard_normal():
    """
    Calculates and plots the standard normal distribution (mean = 0, std = 1).
    """
    x = np.linspace(-4, 4, 1000)  # Range of x-values
    pdf = norm.pdf(x, loc=0, scale=1)  # Calculate PDF for standard normal

    plt.plot(x, pdf, color='blue', linewidth=2)
    plt.title('Standard Normal Distribution (Mean = 0, Std = 1)')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()

# Call the function
plot_standard_normal()
```

## Q14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate Z-score
def calculate_z_score(x, mean, std_dev):
    """
    Calculates the Z-score for a given data point.

    Parameters:
    x : float : Data point
    mean : float : Mean of the dataset
    std_dev : float : Standard deviation of the dataset

    Returns:
    float : Z-score
    """
    return (x - mean) / std_dev

# Example values
data_point = 72
mean = 60
std_dev = 8

# Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point} is: {z:.2f}")

# Plot standard normal distribution
x_vals = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x_vals, loc=0, scale=1)

plt.plot(x_vals, pdf, color='blue', linewidth=2)
plt.title('Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')

# Mark the calculated Z-score
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

## Q15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Hypothetical population parameters
population_mean = 50
population_std = 10

# Sample dataset
sample = [55, 52, 51, 49, 53, 54, 50, 52, 48, 51]
sample_size = len(sample)
sample_mean = np.mean(sample)

# Calculate the Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Z-statistic: {z_stat:.2f}")

# Two-tailed test: Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"P-value: {p_value:.4f}")

# Significance level
alpha = 0.05

# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")
```

## Q16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = [52, 55, 51, 53, 54, 50, 52, 48, 51, 49]

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
sample_size = len(data)

# Confidence level
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

## Q19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
lambda_val = 3  # Average rate of occurrences

# Generate possible values (number of events)
x = np.arange(0, 15)

# Calculate the CDF
cdf = poisson.cdf(x, lambda_val)

# Display CDF values
for i, prob in zip(x, cdf):
    print(f"P(X ≤ {i}) = {prob:.4f}")

# Plot the CDF
plt.step(x, cdf, where='post', color='green', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution (λ = 3)')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

## Q24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    """
    Calculates and prints skewness and kurtosis of the dataset.

    Parameters:
    data : array-like : Dataset
    """
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, this is excess kurtosis (kurtosis - 3)

    print(f"Skewness of the dataset: {skewness:.2f}")
    print(f"Kurtosis of the dataset (Excess Kurtosis): {kurt:.2f}")

# Generate dataset from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate and display skewness and kurtosis
calculate_skewness_kurtosis(data)
```
